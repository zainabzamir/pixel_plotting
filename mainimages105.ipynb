```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Specify the directory containing images
directory = r'deg_60'
# directories = ['deg_60', 'deg_70', 'deg_80', 'deg_110', 'deg_105']

# List to store image objects
images = []
pix_sum = []

# Loop through each file in the directory
for filename in os.listdir(directory):
    # Check if the file is an image (you can add more extensions if needed)
    if filename.endswith(('.tiff')):
        # Construct the full file path
        file_path = os.path.join(directory, filename)
        # Open the image file
        try:
            img = Image.open(file_path)
            # images.append(img)  # Append the image to the list
            img_arr = np.array(img)
            add = np.sum(img_arr)
            pix_sum.append(add)
            
        except Exception as e:
            print(f"Could not load image {filename}: {e}")

# Create a line plot
plt.plot(pix_sum, marker='o')  # `marker='o'` shows dots on each data point

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Plot of List of Numbers')

# Show the plot
plt.show()
```
